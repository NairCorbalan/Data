En el siguiente proyecto analizaremos las ventas de sneakers en varios países, durante un periodo de tiempo determinado (2022).
Las inquietudes que nos acongojan se centran en saber cuáles son las marcas más vendidas, los modelos más vendidos, los países que mayor cantidad de sneakers compra y si hay alguna fecha en especial donde las ventas se hayan disparado.

```python
import pandas as pd
url = 'sneakers_streetwear_sales_data.csv'
df = pd.read_csv(url)
df.head()
```

```python
#mostrar info general del df
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes['Date']
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Amount ($)', palette='viridis')
plt.title('Ventas por region')
plt.xlabel('Country')
plt.ylabel('Amount')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Product Name', y='Amount ($)', palette='magma')
plt.title('Ventas por producto')
plt.xlabel('Product Name')
plt.ylabel('Amount')
```

```python
import matplotlib.pyplot as plt

sns.catplot(data=df, x='Product Name', y='Amount ($)', col='Country', kind='bar', height=5, aspect=0.8, palette='viridis', errorbar=None)
plt.suptitle('Ventas por producto y region', y=1.02)
plt.show()
```
